```python
# Encoding Categorical Data
df=pd.get_dummies(df,drop_first=True)
```

```python
# Splitting x and y
x=df.drop("math score",axis=1)
y=df["math score"]
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into Training set and Test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=37)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Application of Linear Regression model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
# Predicting the Test results
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
# Calculating the Mean absolute error, Mean squared error, Root mean squared error and R2 score
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print("MAE :", mean_absolute_error(y_test,y_pred))
print("MSE :", mean_squared_error(y_test,y_pred))
print("RMSE :", np.sqrt(mean_squared_error(y_test,y_pred)))
print("R2 Score :", r2_score(y_test, y_pred))
```

```python
# Application of Decision Tree model
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=37)
dt.fit(x_train,y_train)
```

```python
# Predicting the Test results
dt_pred=dt.predict(x_test)
```

```python
dt_pred
```

```python
# Calculating the R2 score
print("Decision Tree R2 Score:",r2_score(y_test,dt_pred))
```

```python
# Application of Random Forest model
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(random_state=37)
rf.fit(x_train, y_train)
```

```python
# Predicting the Test results
rf_pred=rf.predict(x_test)
```

```python
rf_pred
```

```python
# Calculating the R2 score
print("Random Forest R2 Score:",r2_score(y_test,rf_pred))
```

```python
# Comnparing the Models
comparison=pd.DataFrame({"Model":["Linear Regression","Decision Tree","Random Forest"],
        "R2 Score":[r2_score(y_test,y_pred),r2_score(y_test,dt_pred),r2_score(y_test,rf_pred)]})
comparison
```

```python
# Best Model
comparison.sort_values(by="R2 Score",ascending=False)
```

## The Best model is Linear Regression model for the Student Performance dataset.
