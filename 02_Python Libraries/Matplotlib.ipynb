# Matplotlib

Used to represent the data in a visually appealing manner in the form of line charts, bar graphs, pie charts, histograms, sactter plots, area plots, heatmaps, etc.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Line Chart

```python
l1=[3,7,10,37,77]
l2=[10,20,30,40,50]
plt.plot(l1,l2,marker="o")
plt.show()
```

```python
help(plt.plot)
```

## Bar Graph

```python
plt.bar(l1,l2,color=["blue","yellow","green","red","pink"],width=2)
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales Portfolio")
plt.show()
```

```python
help(plt.bar)
```

## Histogram

```python
a1=[10,20,30,40,50]
plt.hist(a1)
plt.show()
```

```python
help(plt.hist)
```

## Scatter Plots

```python
plt.scatter(l1,l2,color="Lightblue")
plt.show()
```

```python
help(plt.scatter)
```

## Stacked Plots

One bar graph over another bar graph, stacking happening in the bar graphs.

```python
var1=["HS","DK","DP","GK","AS","JS"]
y1=[37,10,45,60,34,20]
y2=[5,10,15,20,25,30]
plt.bar(var1,y1,color="blue")
plt.bar(var1,y2,color="green")
plt.show()
```

## Pie Charts

```python
x=[10,37,7,3,43]
label=["HS","DK","DP","GK","AS"]
plt.pie(x,labels=label)
plt.legend()
plt.show()
```

```python
help(plt.pie)
```

```python
plt.pie(x,labels=label,explode=[0.5,0,0,0,0])
plt.show()
```
