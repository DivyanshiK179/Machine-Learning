## 1-Paired T Test

We Prove: The expected value(mean) of a sample of independent observation is equal to the given population mean. 

```python
import pandas as pd
```

```python
df=pd.read_csv("D:/ML Internship/Production_Charge_lyst1729241054998.csv")
df
```

```python
df=df.drop("Unnamed: 0",axis=1,errors="ignore")
df
```

```python
import seaborn as sns
```

```python
#Check for SND
sns.distplot(df["production_cost"],color="red",hist=0)
```

```python
#Null hypothesis ans Alternate hypothesis
#H0:U=500
#H1:U!=500
```

```python
from scipy import stats
```

```python
stats.ttest_1samp(df["production_cost"],500)
```
