# Task 2: Decision Tree, Random Forest & Ensemble Techniques

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
x=iris.data
y=iris.target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=37)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
dt=DecisionTreeClassifier(random_state=37)
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("Decision Tree Accuracy:",accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
rf=RandomForestClassifier(n_estimators=100,random_state=37)
rf.fit(x_train, y_train)
```

```python
y_pred_rf=rf.predict(x_test)
y_pred_rf
```

```python
print("Random Forest Accuracy:",accuracy_score(y_test, y_pred_rf))
```

```python
gb=GradientBoostingClassifier(random_state=37)
gb.fit(x_train, y_train)
```

```python
y_pred_gb=gb.predict(x_test)
y_pred_gb
```

```python
print("Gradient Boosting Accuracy:",accuracy_score(y_test,y_pred_gb))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
plot_tree(dt,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
plt.title("Decision Tree on Iris Dataset")
plt.show()
```

# Part B: Unsupervised Learning with K-Means

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=3,random_state=37,n_init=10)
clusters=km.fit_predict(x)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(7,5))
plt.scatter(x[:,0],x[:,1],c=clusters,cmap="viridis")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("K-Means Clustering (k=3)")
plt.show()
```

Supervised learning uses labeled data to train a model and predict known output classes. Algorithms such as KNN, SVM, and Naive Bayes learn from input-output pairs and are mainly used for classification and prediction tasks. Unsupervised learning uses unlabeled data and finds hidden patterns or groups without knowing the correct answers. K-Means is an unsupervised algorithm that clusters similar data points into groups based on their features. Supervised learning generally provides higher prediction accuracy when labeled data is available, while unsupervised learning is useful for discovering patterns, customer segmentation, and exploratory data analysis.
