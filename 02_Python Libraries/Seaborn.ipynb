# Seaborn

Used for data visualisation for complex data structures and complex type of graphs.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("C:/Users/Divyanshi Kesharwani/Downloads/Data.csv")
print(df)
```

### Univariate Categorical Analysis

```python
df["Country"].value_counts()
sns.catplot(x="Country",data=df,kind="count")
```

```python
help(sns.catplot)
```

### Univariate Numerical Analysis

```python
sns.distplot(df["Salary"])
```

```python
sns.distplot(df["Salary"],hist=False,color="green")
```

```python
sns.histplot(df["Salary"])
```

```python
plt.figure(figsize=(5,2))
sns.histplot(df["Salary"])
```

```python
sns.jointplot(x="Salary",y="Age",data=df,kind="scatter",hue="Country")
```

### Bivariate Categorical Analysis

```python
sns.boxplot(x="Country",y="Purchased",data=df)
```

```python
sns.catplot(x="Country",y="Purchased",data=df,kind="swarm")
```

```python
sns.catplot(x="Country",y="Purchased",data=df,kind="violin")
```
