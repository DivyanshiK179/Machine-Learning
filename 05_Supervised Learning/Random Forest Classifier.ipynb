```python
import pandas as pd
```

```python
df=pd.read_csv("D:/ML Internship/car_evaluation_lyst1729241367952.csv")
df
```

```python
#Rename the columns
col_name=["Buying","Maintenace","Doors","Persons","Lugboot","Safety","Class"]
df.columns
```

```python
df.columns=col_name
```

```python
df.describe()
```

```python
#Putting the value of x and y
x=df.drop("Class",axis=1)
y=df["Class"]
```

```python
x
```

```python
y
```

```python
#Splitting the dataset into Training and Test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=37)
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#Encoding of categorical data: Preprocessing of data
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
x_train=encoder.fit_transform(x_train)
x_test=encoder.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
x_train.dtype
```

```python
#Application of model
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=100,random_state=77,bootstrap=True)
rfc.fit(x_train,y_train)
```

```python
y_pred=rfc.predict(x_test)
y_pred
```

```python
y_test
```

```python
#Classification report, Accuracy score and Confusion matrix
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_pred,y_test)
```
