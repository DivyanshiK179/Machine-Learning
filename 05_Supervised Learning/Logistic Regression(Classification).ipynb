```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df=sns.load_dataset("iris")
df.head()
```

```python
df
```

```python
#To check whether the species column has binary classification or not
df["species"].value_counts()
```

```python
#To find the number of null enteries
df.isnull().sum()
```

```python
#To make the data only of binary classification
df=df[df["species"]!="setosa"]
df
```

```python
df["species"].value_counts()
```

```python
#Mapping of binary classification as 0 & 1
df["species"]=df["species"].map({"versicolor":0,"virginica":1})
df
```

```python
#Splitting of dataset(dependent and independent features)
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
#Splitting the dataset into Train,ing and Test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#Application of model into the dataset
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
#Checking the accuracy, classification report and confusion matrix
from sklearn.metrics import accuracy_score,classification_report, confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_pred,y_test)
```
