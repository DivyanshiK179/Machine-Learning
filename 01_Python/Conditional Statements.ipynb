# Conditional Statements

Conditional statements are the part of the python program which we use whenever we want the block of code to get executed on the basis of some condition.

Indentation is the 4 space or 1 tab gap which by default gets picked up from the newline whenever we end any previous python statement from colon. 

Syntax:
if(condition1):
    block of code
    .......
elif(condition2):
    block of code
    .......
else:
    block of code
    .......

```python
num=37
if(num%2==0):
    print("Divisible by 2.")
else:
    print("Not divisible by 2.")
```

```python
num2=3
if(num2==3):
    print("Equal")
```

```python
num3=0
if(num3>0):
    print("Positive")
elif(num3<0):
    print("Negative")
else:
    print("Zero")
```
