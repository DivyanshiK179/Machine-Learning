```python
import math
from scipy import stats
```

```python
def z_test(population_mean,population_std,sample):
    z=(sample.mean()-population_mean)/(population_std/math.sqrt(len(sample)))
    return z,stats.norm.cdf(z)
```

```python
z_test(30,12.5,df["Processing_time"])
```
