# Data Types in Python

## Primitive Data Types

It is homogenous in nature and have the property that one indentifier captures only one value.
1. int
2. float
3. str
4. bool
5. None

```python
num_1=37
print(num_1)
print(type(num_1))
```

```python
x=37.10
print(x)
print(type(x))
```

```python
str_1="Harsh is a good boy!"
print(str_1)
print(type(str_1))
```

```python
var=True
print(var)
print(type(var))
```

```python
y=None
print(y)
print(type(y))
```

# Non-Primitive Data Types

It is non homogenous in nature and have the property that one identifier can capture collection of different values.

1. List
2. Tuple
3. Set
4. Dictionary


### List

```python
l_1=[3,7,10,37,77]
```

```python
l_2=[3,7,37.10,"Harsh",True,None]
```

```python
print(l_2)
```

```python
print(type(l_1))
```

1. List is always declared inside [].
2. List follows indexing, both backward and forward.
3. Lists are mutable in nature.
4. Lists follows slicing.

```python
print(l_1[3])
```

```python
help(list)
```

```python
l_2[4]=False
```

```python
print(l_2)
```

### Tuple

```python
t_1=(3,7,10,37,77)
```

```python
t_2=(3,7,37.10,"Harsh",True,False,[3,7])
```

```python
print(t_2)
```

```python
print(type(t_1))
```

1. Tuple is always declared inside ().
2. We can declare any data type inside the tuple.
3. Tuple follows indexing, both forward and backward.
4. Tuples follows slicing (happens only sequentially).
5. Tuples are immutable in nature.

```python
t_1[::-1]
```

```python
help(tuple)
```

```python
print(t_2[-1:-3:-1])
```

### Sets

```python
s_1={3,7,10,37,77}
```

```python
s_2={3,7,37.10,"Harsh",True,None}
```

```python
print(s_2)
```

```python
print(type(s_1))
```

1. Sets are unordered in nature.
2. Sets do not follow indexing.
3. Sets elememts are unique in nature, no duplicate of element is allowed.
4. When we want to declare an empty set, then we need to use built-in function set_empty().
Similarly:list_empty(), tuple_empty().
5. Sets are m utable in nature and it accepts only primitive data types along with tuple in it.

```python
s_3={3,3,7,10,37,37}
```

```python
print(s_3)
```

```python
help(set)
```

```python
list_empty=[]
print(type(list_empty))
```

```python
tuple_empty=()
print(type(tuple_empty))
```

```python
set_empty={}
print(type(set_empty))
```

### Dictionary

```python
d_1={3:"HS",7:"DK"}
```

```python
print(d_1)
```

```python
d_2={"Name":"DK","Age":20,"Course":"BTECH"}
```

```python
print(type(d_2))
```

```python
d_2.keys()
```

```python
d_1.values()
```

1. Dictionary is a combination of key-value pairs.
2. Keys only accept primitive data types and tuples in it.
3. Keys are immutable in nature.
4. Keys are always unique.
5. Values can be of any data type.
6. Values are mutable in nature.
7. Values can be duplicated.
8. Dictionaries are mutable in nature.

```python
d_1[3]
```

```python
d_2["Section"]="B1"
```

```python
print(d_2)
```

```python
help(dict)
```
