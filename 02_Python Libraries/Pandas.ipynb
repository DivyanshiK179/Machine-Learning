# Pandas

1. Full form: Panel Data
2. Very helpful in performing data analysis.
3. 1-D data set is known as Series.
4. 2-D data set is known as Dataframe.

```python
import pandas as pd
s1=pd.Series(data=[3,7,77],index=["HS","DK","SG"])
print(s1)
```

```python
s2=pd.Series(data=[10,20,30],index=["Physics","Chemistry","Maths"])
print(s2)
```

```python
d1=pd.DataFrame(data={"HS":[3,37],"DK":[7,10]})
print(d1)
```

```python
d2=pd.DataFrame(data={"HS":[3,37],"DK":[7,10]},index=["a","b"])
print(d2)
```

```python
# Both the series should have same indices
d3=pd.DataFrame(data={"col1":s1,"col2":s2})
print(d3)
```
