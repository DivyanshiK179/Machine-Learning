# Operators

## Types of Operators

1. Arithmetic Operators
2. Assignment Operators
3. Relational Operators
4. Logical Operators
5. Bitwise Operators
6. Membership Operators
7. Identity Operators

### Arithmetic Operators

+, -, *, /, **, %, //

```python
a=3
b=7
print(a**b)
```

### Assignment Operators

=, +=, -=, *=, /=, %=, //=, **=

```python
var=37
var+=10
print(var)
```

### Relational Operators

==, >=, <=, !=, >, <

```python
h=37
d=10
print(h>d)
```

```python
print(h!=d)
```

### Logical Operators

and, or, not

```python
x=37
y=10
print(x%2!=0 and y==10)
```

```python
print(x%2==0 or y==10)
```

```python
print(not(y))
```

### Bitwise Operators

&, |, ~

### Membership Operators

in, not in

```python
l_1=[3,7,10,37,77]
print(37 in l_1)
```

```python
print(77 not in l_1)
```

### Identity Operators

is, is not

```python
a=10
b=10
print(id(a))
```

```python
print(id(b))
```

```python
print(a is b)
```

```python
c=37
print(a is not c)
```
