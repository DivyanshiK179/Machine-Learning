# Functions

1. Functions are the block of code which only runs when it is called.
2. Once declared, we can use it again and again.
3. They are flexible in nature (if declared in one jupyter notebook then can be called in another notebook also).
4. Functions are universal in nature.

Syntax:
1.Function Definition:
def function_name(parameter(s)):
    body of function
    .......
    print()/return

2.Function Calling:
function_calling(argument(s))

## Types of Functions

### Built-in Functions

They are already declared in python and we just need to call them to use.
print(), type() are some examples.

### User-Defined Functions

Functions that are created by the user for their ease of code and reusability.

### Anonymous Functions

1. Functions that do not have any sort of name.
2. Their syntax is different from usual syntax of functions.
3. We usually create them in a single line.
4. You can give only one expression for a single lambda function.

### Recursive Functions

Function calling itself is called a recursive function.

```python
def summ(num1,num2):
    return num1+num2

n1=int(input("Enter number1:"))
n2=int(input("Enter number2:"))
print("Sum of numbers:",summ(n1,n2))
```

```python
summ(3,7)
```

```python
def sub(num1,num2):
    print("Difference of numbers:",num1-num2)

n1=int(input("Enter number1(larger one):"))
n2=int(input("Enter number2:"))
sub(n1,n2)
```

```python
sub(7,3)
```

```python
def master(a,b):
    ans_1=summ(a,b)
    ans_2=sub(a,b)
    print(ans_1,ans_2)
master(37,10)
```

For permanent output, use return keyword.

## Variable Length Arguments

```python
def sum_n(*n):
    summ=0
    for i in n:
        summ+=i
    return summ
    
sum_n(3,7,10,37,77)
```

```python
sum_n(10,20,30,40,50)
```

## Anonymous Functions

Syntax:
1. Function Definition:
variable_name=lambda(parameter(s)):expression
2. Function Calling:
3. variable_name(argument(s))

```python
cube=lambda a:a**3
cube(3)
```

```python
cube(37)
```

```python
e_o=lambda n:"Even number" if n%2==0 else "Odd number"
e_o(37)
```

```python
e_o(10)
```

## Recursive Functions

Syntax:
1. Function Definition:
   def function_name(parameter(s)):
       body of function
       function_name(argument(s))
2. Function Calling:
   function_name(argument(s))

```python
def fact(n):
    if(n==1):
        return 1
    else:
        return n*fact(n-1)

fact(7)
```

```python
fact(3)
```
