```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv("D:/ML Internship/placement_lyst1729240397995.csv")
df
```

```python
# Plotting the graph
sns.scatterplot(x="cgpa",y="placement_exam_marks",data=df,hue="placed")
```

```python
# Splitting the value of x and y
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
# Splitting the data into Training set and Test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Application of model
from sklearn.linear_model import Perceptron
pr=Perceptron()
pr.fit(x_train,y_train)
```

```python
pr.score(x_train,y_train)
```

```python
pr.score(x_test,y_test)
```

```python
# Graph to see the behaviour of model
plt.figure(figsize=(10,5))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=pr)
```
