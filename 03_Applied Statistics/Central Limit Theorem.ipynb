```python
import pandas as pd
```

```python
df=pd.read_csv("D:/ML Internship/employee_lyst1729240478122.csv")
df
```

```python
population=df["Base Salary"]
population.describe()
```

```python
import seaborn as sns
```

```python
sns.distplot(population,color="green")
```

```python
mean=[]
def cal_sample_mean(sample_size,no_of_sample_mean):
    for i in range(no_of_sample_mean):
        sample_base_salary=population.sample(n=sample_size)
        sample_mean=sample_base_salary.mean()
        mean.append(sample_mean)
    return mean
```

```python
mean_1=cal_sample_mean(3,500) 
mean_1
```

```python
sns.distplot(mean_1,color="red")
# Central limit theorem ism proved.
```
