```python
Syntax:
try:
    block of code
    # Line of Code which has potential to cause error
except:
    optional block
    # Handling of Exceptions(if required)
finally:
    block of code
    # Always executed
```

1. Finaaly block is always executed after leaving the try block.
2. In case if some exception is not handled by except block, it is re-raised after execution of finally block.
3. Finally block is used to deallocate the system resources.

```python
try:
    n1=int(input("Enter number1:"))
    n2=int(input("Enter number2:"))
    ans=n1/n2
    print(ans)
except ZeroDivisionError:
    print("Please enter the number other than 0, you are dividing the number by 0!")
except ValueError:
    print("Please enter the int type value!")
finally:
    print("End of program!")
```
