```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
from sklearn.datasets import load_breast_cancer
cancer_data=load_breast_cancer()
cancer_data.keys()
```

```python
print(cancer_data.DESCR)
```

```python
df=pd.DataFrame(cancer_data["data"],columns=cancer_data["feature_names"])
df
```

```python
df.columns
```

```python
# Feature Scaling of the data
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
scaled_data
```

```python
# Dimensionality Reduction by applying PCA Algorithm
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
```

```python
data_pca=pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
# Variance of the two columns: Variance of 1 is always greater than variance of 2 and so on.
pca.explained_variance_
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_data["target"],cmap="plasma")
plt.xlabel("First PC")
plt.ylabel("Second PC")
```
