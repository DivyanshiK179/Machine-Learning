# Iterative Statements

## For Loops

1. We know the number of iterations before.
2. We use iterative data types:
   a] List
   b] Tuple
   c] Set
   d] Dictionary
   e] String

Syntax:
for temp_var in iterable:
    body of for loop

Built-in functions:
    a] range()
    b] enumerate(): Iterable elements with their index (index, element)
    c] zip(): To give more than one iterative sequences at once

```python
l_1=[3,7,10,37,77]
for i in l_1:
    print(i)
```

```python
s_1={3,3,7,10,37,37,77}
for i in s_1:
    print(i)
```

```python
d_1={3:"HS",7:"DK",77:"SG"}
for i in d_1:
    print(i)
```

```python
for j in d_1.values():
    print(j)
```

```python
for k in d_1.items():
    print(k)
```

```python
for i in range(1,11):
    print(i)
```

```python
for i in enumerate(l_1):
    print(i)
```

```python
l_2=[10,20,30,40,50]
for i,j in zip(l_1,l_2):
    print(i+j)
    
```

## While Loops

1. We do not know the number of iterations before.
2. We use  iterative data types.

Syntax:
while(condition):
    body of while loop
    one statement to make the while loop false




```python
h=3
while(h<=10):
    print(h)
    h+=1
```

```python
num=int(input("Enter a number:"))
while(num!=0):
    print(num)
    num=int(input("Enter the next number:"))
print("End")
```

# Control Flow Statements

## Pass

1. Pass is the NULL statement that helps us to avoid errors.
2. It reserves the space for future use and references.

```python
for i in range(1,11):
    if(i==6):
        pass
    print(i)
```

## Break

Helps to terminate the loop before completition.

```python
for i in range(1,11):
    if(i==6):
        break
    print(i)
```

## Continue

Helps us to skip the current iteration of the loop and move to the next iteration.

```python
for i in range(1,11):
    if(i==6):
        continue
    print(i)
```
