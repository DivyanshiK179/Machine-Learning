# Univariate Feature Imputation

```python
import numpy as np
x=np.array([[np.nan,2],[6,np.nan],[10,37]])
x
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy="mean")
imp.fit_transform(x)
```

```python
# For categorical features
import pandas as pd
df=pd.DataFrame([["h","s"],[np.nan,"k"],["d",np.nan],["g","k"]],dtype="category")
df
```

imp_1=SimpleImputer(strategy="most_frequent")
print(imp_1.fit_transform(df))

# Multivariate Feature Imputation

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp_it=IterativeImputer(max_iter=10,random_state=7)
imp_it.fit_transform(x)
```

```python
# K-Nearest neighbour imputer
y=np.array([[1,2,np.nan],[10,37,np.nan],[np.nan,6,5],[8,8,7]])
y
```

```python
from sklearn.impute import KNNImputer
imputer=KNNImputer(n_neighbors=2,weights="uniform")
imputer.fit_transform(y)
```
