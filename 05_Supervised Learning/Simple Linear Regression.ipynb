```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("D:/ML Internship/Company_Profit_lyst1729240579399.csv")
df
```

```python
#Check whether there is relationship between dependent and independent feature(linear relationship)
sns.pairplot(df)
```

```python
#Separate the input and target feature
x=df.iloc[:,:-1].values
y=df.iloc[:,1:].values
```

```python
x
```

```python
y
```

```python
#Splitting of dataset into train  & test part
x=x.reshape(-1,1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#Appliaction of model
from sklearn.linear_model import LinearRegression
linear_l=LinearRegression()
linear_l.fit(x_train,y_train)
y_pred=linear_l.predict(x_test)
```

```python
#Score of the model
linear_l.score(x_train,y_train)
```

```python
#Finding of error
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test,y_pred)
```

```python
#Finding the coefficient
linear_l.coef_
```

```python
linear_l.intercept_
```
