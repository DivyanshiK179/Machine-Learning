```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson
```

# Poisson Distribution

```python
lam=5 #average rate or constant mean
k=np.arange(0,15) #possibvle value of k or t

pmf_value=poisson.pmf(k,lam)
pmf_value
```

```python
plt.bar(k,pmf_value,color="green",alpha=0.7)
plt.title("Poisson Distribution")
plt.xlabel("Number of events")
plt.ylabel("Probability")
plt.grid(axis="y")
plt.show()
```

# Exponential Distribution

```python
from scipy.stats import expon
```

```python
lam_v=1
x=np.linspace(0,10,100) # generate the value of x
x

pdf_value=expon.pdf(x,scale=1/lam_v)
pdf_value
```

```python
plt.plot(x,pdf_value,color="red")
plt.title("Exponential Distribution")
plt.xlabel("Samples")
plt.ylabel("Probability")
plt.grid()
plt.show()
```
