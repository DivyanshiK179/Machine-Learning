# Standard Normal Distribution

```python
import pandas as pd
```

```python
df=pd.read_csv("D:/ML Internship/BEML_lyst1729240526796.csv")
df
```

```python
df=df[["Date","Close"]]
df
```

```python
df=df.set_index(pd.DatetimeIndex(df["Date"]))
df
```

```python
df=df.drop("Date",axis=1)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(df["Close"])
plt.xlabel("Year")
plt.ylabel("Close")
```

```python
df["Gain"]=df["Close"].pct_change(periods=1)
df
```

```python
sns.distplot(df["Gain"])
```

```python
df["Gain"].describe()
```

# Confidence Interval

```python
from scipy import stats
```

```python
beml_gain_ci=stats.norm.interval(0.95,loc=df["Gain"].mean(),scale=df["Gain"].std())
beml_gain_ci
```
