## Different ways of creating NumPy Arrays

```python
import numpy as np
```

```python
var1=np.array([1,2,3,4])
print(var1)
```

```python
var2=np.array([[1,2,3],[4,5,6]])
print(var2)
```

```python
np.zeros([2,3])
```

```python
np.zeros([3,3],dtype="int")
```

```python
np.ones([4,3])
```

```python
var3=np.ones([3,4],dtype="int")
print(var3)
```

```python
np.random.rand(4,4)
```

```python
r=np.random.rand(3,4)
print(r)
```

```python
np.random.seed(3)
r2=np.random.rand(3,4)
print(r2)
```

```python
np.random.randn(4,4)
```

```python
np.random.seed(4)
r2=np.random.randint(low=50,high=100,size=(3,4))
print(r2)
```

## Slicing and Indexing in Numpy Arrays

```python
print(r2[1,1])
```

```python
print(r2[2,3])
```

```python
print(r2[1:,1:])
```

```python
print(r2[0:1,0:2])
```

```python
print(r2[1:2,1:])
```

## Stacking and Splitting of a Matrix

```python
ar2=np.arange(1,21).reshape(4,5)
ar3=np.arange(21,41).reshape(4,5)
print(ar2)
print(ar3)
```

```python
np.vstack((ar2,ar3))
```

```python
np.hstack((ar2,ar3))
```

```python
np.vsplit(ar2,2)
```

```python
np.hsplit(ar1,2)
```
