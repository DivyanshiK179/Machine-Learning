```python
import pandas as pd
```

```python
df=pd.read_csv("D:/ML Internship/Titanic-Dataset.csv")
df
```

```python
df.columns
```

```python
df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis=1,inplace=True)
df
```

```python
#Picking the value of x and y
x=df.drop("Survived",axis=1)
y=df["Survived"]
```

```python
x
```

```python
y
```

```python
#Encoding of Categorical data
dummy=pd.get_dummies(x["Sex"],dtype="int")
dummy
```

```python
#Concatenating the dummy variables
x=pd.concat([x,dummy],axis=1)
```

```python
x
```

```python
x.drop(["Sex"],axis=1,inplace=True)
x
```

```python
x.dtypes
```

```python
#Checking and filling of null values
x.isnull().sum()
```

```python
x["Age"]=x["Age"].fillna(x["Age"].median())
```

```python
x.isnull().sum()
```

```python
#Splitting of dataset into Training set and Test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#Application of model
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
#Classification report
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
#Confusion matrix
confusion_matrix(y_test,y_pred)
```

```python
#The probability behind the x features
pd.options.display.float_format="{:,.1f}".format
x=model.predict_proba(x_test[:10])
```
